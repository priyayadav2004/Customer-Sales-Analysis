we have to remove the no. from the alphabet . for this we import the regex pattern 

```python
import re

def extract_Age(Age):
    Age_num = re.findall('[0-9]+', str(Age))
    if len(Age_num) > 0:
        return Age_num[0]
    else:
        return Age

df['Age'] = df['Age'].apply(lambda x: extract_Age(x))
    

```

handle the outlires using inter quartile range (IQR) since we have 3 numerical columns so we will see in all 3 columns 

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

plt.figure(figsize=(15,5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f"{col} Boxplot")

plt.tight_layout()
plt.show()
```

we remove the negative values is was neccesary for removing the outlires

SciPy = Scientific Python library
used for advace calculation

```python
from scipy import stats
stats.zscore(df[['Age','Purchase_Amount']])
```

conver the negative values into positive 

```python
z_score = np.abs(stats.zscore(df[['Age','Purchase_Amount']]))
```

```python
df[( z_score > 3 ).any (axis = 1 )]
```

```python
df_clean = df[~( z_score > 3 ).any (axis = 1 )]
```

```python
df_clean.shape
```

```python
cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
df.to_csv('clean_sales_data.csv', index=False)
```
